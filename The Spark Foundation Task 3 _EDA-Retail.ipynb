# Exloratory Data Analysis

```python
cor = df.corr()
cor
```

```python
#from below graph we can see that profit and Sales shows Positive correlation 
plt.subplots(figsize = (5,5))    
sns.heatmap(cor, annot = True)
```

```python
#showcase pairwise relationships between different variables in the DataFrame
#Below i have used Pairplot from seaborn based on Category and Region
sns.pairplot(df,hue = 'Category',palette = 'rainbow')  
```

```python
#Plotting the KDE Plot
plt.figure(figsize = (8,5))
sns.kdeplot(df['Sales'],color='blue',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='red',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.legend()

#from below graph we can see profit is more than sale 
#but we can also see there are some areas where Profit could be increased
```

```python
# The above Graph shows that , count of Customers
#The customers from California state are high in number and from Wyoming are low in number
#From CALIFORNIA State the supermatket gains more profit and from WYOMING gains lower profit


plt.figure(figsize=(10,8))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
df_state = df.groupby(['State'])[['Sales','Discount','Profit']].mean()
df_state.plot(kind='bar',subplots= True,figsize=(20,22))
plt.title('State wise analysis of Sale',fontsize=25)
#Highest Amount of Sales - Wyoming
#Highest Discount - illinois
#Highest Profit - Vermont
```

```python
df_city = df.groupby(['City'])[['Sales','Discount','Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head(15)
df_city = df_city.iloc[:15]
```

```python
df_city.plot(kind='bar',subplots= True,figsize=(20,22))
plt.title('City wise analysis of Sale',fontsize=25)
```

```python
#df_Category will contain the sales, discount, and profit values for each 'Category' in the original DataFrame df
df_Category = df.groupby(['Category'])[['Sales','Discount','Profit']].mean()
df_Category.head()
```

```python
#This code would generate multiple bar plots arranged in a grid, each representing the mean values of 'Sales', 'Discount', and 'Profit' for each 'Category'.
#The title 'City wise analysis of Sale' would be set at the top of the entire plot, indicating the context of the analysis.
#Maximum Sales and Profit obtain in Technology
#Minimum Profit Obtained in Furniture
df_Category.plot(kind='bar',subplots= True,figsize=(20,22))
plt.title('City wise analysis of Sale',fontsize=25)
```

```python
#Here i have  created a new DataFrame (df_cat) that provides the average sales, discount, and profit values for each 'Sub-Category' present in the original DataFrame df
df_cat= df.groupby(['Sub-Category'])[['Sales','Discount','Profit']].mean()
df_cat.head()
```

```python
#This code snippet would create multiple bar plots, each representing the mean sales, discount, and profit values for different sub-categories, arranged in a grid.
# Below graph we can see Copies have High Sales and Profit
#Binders, Machines and Tables have high Discounts
df_cat.plot(kind='bar',subplots= True,figsize=(20,22))
plt.title('City wise analysis of Sale',fontsize=25)
```

```python
#This code creates a new DataFrame df_Region that provides the average sales, discount, and profit values for each 'Region' present in the original DataFrame df. 
df_Region = df.groupby(['Region'])[['Sales','Discount','Profit']].mean()
df_Region.head()
```

```python
#This code will generate individual bar plots for 'Sales', 'Discount', and 'Profit', showing the average values across different regions. 
#Profit is highest in the West Region
df_Region.plot(kind='bar',subplots= True,figsize=(20,22))
plt.title('City wise analysis of Sale',fontsize=25)
```
